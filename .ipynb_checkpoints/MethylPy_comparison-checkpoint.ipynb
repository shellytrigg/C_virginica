```python
#run methylpy with bowtie2 alignment
```

```python
%%bash

find /Volumes/web/seashell/bu-serine-wd/18-04-07/zr2096_*R1* \
| xargs basename -s _s1_R1.fastq.gz | \
xargs -I{} /Users/Shelly/anaconda3/bin/methylpy paired-end-pipeline \
--read1-files /Volumes/web/seashell/bu-serine-wd/18-04-07/{}_s1_R1.fastq.gz \
--read2-files /Volumes/web/seashell/bu-serine-wd/18-04-07/{}_s1_R2.fastq.gz \
--sample {}_Cvirgv300_pe_bt2 \
--path-to-output /Volumes/web/metacarcinus/Cvirginica/methylpy/ \
--forward-ref /Volumes/web/metacarcinus/Cvirginica/GENOMES/v300/methylpy_genome/methylpy_bt_f \
--reverse-ref /Volumes/web/metacarcinus/Cvirginica/GENOMES/v300/methylpy_genome/methylpy_bt_r \
--ref-fasta /Volumes/web/metacarcinus/Cvirginica/GENOMES/v300/Bismark_genome/Cvirginica_v300.fa \
--num-procs 2 \
--aligner bowtie2 \
--aligner-options "--score-min L,0,-1.2 -u 100000 -I 60 --no-mixed --no-discordant" \
--remove-clonal True \
--path-to-picard /Users/Shelly/anaconda3/bin/ \
--path-to-samtools /Users/Shelly/anaconda3/bin/ \
--add-snp-info True \
--binom-test True \
--unmethylated-control 0.005
    
    
```
