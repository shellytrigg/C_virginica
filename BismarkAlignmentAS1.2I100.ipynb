```python
%%bash

find /Volumes/web/seashell/bu-serine-wd/18-04-07/zr2096_*R1* \
| xargs basename -s _s1_R1.fastq.gz | xargs -I{} /gscratch/srlab/programs/Bismark-0.19.0/bismark \
--genome /Volumes/web/metacarcinus/Cvirginica/GENOMES/v300/Bismark_genome/ \
--path_to_bowtie /gscratch/srlab/programs/bowtie2-2.3.4.1-linux-x86_64
--score_min L,0,-1.2 \
-I 100 \
--non_directional \
-1 /Volumes/web/seashell/bu-serine-wd/18-04-07/{}_s1_R1.fastq.gz \
-2 /Volumes/web/seashell/bu-serine-wd/18-04-07/{}_s1_R2.fastq.gz \
-o /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt7/ \
--samtools_path /gscratch/srlab/programs/samtools-1.9/




```

```python
#run deduplication
```

```python
%%bash
/gscratch/srlab/programs/Bismark-0.19.0/deduplicate_bismark \
--bam -p \
/Volumes/web/metacarcinus/Cvirginica/Bismark_attempt7/*.bam \
-o /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt7/ \
2> /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt7/dedup.err \
--samtools_path /gscratch/srlab/programs/samtools-1.9/
```
