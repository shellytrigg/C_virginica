copy genome to working directory folder

```python
%%bash
cp /Volumes/web-1/halfshell/genomic-databank/Cvirginica_v300.fa /Volumes/web/metacarcinus/Cvirginica/GENOMES/v300/Bismark_genome/
```

run bismark genome preparation

```python
!bismark_genome_preparation --path_to_bowtie /Users/Shelly/anaconda3/bin/ --verbose /Volumes/web/metacarcinus/Cvirginica/GENOMES/v300/Bismark_genome/
```

Then I ran bismark subsetting only the first 10K reads from the first two .fastqs, specifying reads are non-directional, and using a minimum score threshold that Steven previously used

```python
!bismark -u 100000 --non_directional --score_min L,0,-0.6 \
--genome /Volumes/web/metacarcinus/Cvirginica/GENOMES/v300/Bismark_genome/ \
-1 /Volumes/web/seashell/bu-serine-wd/18-04-07/zr2096_1_s1_R1.fastq.gz \
-2 /Volumes/web/seashell/bu-serine-wd/18-04-07/zr2096_1_s1_R2.fastq.gz \
-o /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/ 
```

run Bismark Methylation extractor

```python
!bismark_methylation_extractor --gzip -bedGraph /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/zr2096_1_s1_R1_bismark_bt2_pe.bam -o /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/ 
```

generate Bismark Nucleotide Coverage report

```python
!bam2nuc /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/zr2096_1_s1_R1_bismark_bt2_pe.bam\
--dir /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/ \
--samtools /Users/Shelly/anaconda3/bin/samtools \
--genome_folder /Volumes/web-1/halfshell/genomic-databank/

```

generate Bismark report

```python
!bismark2report \
--alignment_report /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/zr2096_1_s1_R1_bismark_bt2_PE_report.txt \
--splitting_report /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/zr2096_1_s1_R1_bismark_bt2_pe_splitting_report.txt \
--mbias_report /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/zr2096_1_s1_R1_bismark_bt2_pe.M-bias.txt \
--dir /Volumes/web/metacarcinus/Cvirginica/Bismark_attempt1/ 
```
